```python
cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == "object"):
        cat.append(i)
        x = X[i].mode()[0]
        X[i] = X[i].fillna(x)
    else:
        con.append(i)
        x = X[i].mean()
        X[i] = X[i].fillna(x)
```
